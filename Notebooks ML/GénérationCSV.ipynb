```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.layers import Input, Dense, Dropout
from keras.models import Model

# Load song data into a DataFrame
songs_df = pd.read_csv('/content/gdrive/MyDrive/songinfos40sec.csv')

# Extract the features you want to use for song recommendations
song_features = songs_df[['zero_cross_sum', 'zero_cross_rate_mean','mfcc_mean_globalmfcc1_mean']]


# Split the data into a training and test set
train_df, test_df = train_test_split(song_features, test_size=0.2)

# Define the input layer for the features
features_input = Input(shape=(3,)) # 3 is the number of features zcr, chroma, mfcc

# Define the dense layers
x = Dense(64, activation='relu')(features_input)
x = Dropout(0.1)(x)
x = Dense(32, activation='relu')(x)
x = Dropout(0.1)(x)
x = Dense(16, activation='relu')(x)

# Define the output layer
output = Dense(1, activation='sigmoid')(x)

# Create the model
model = Model(inputs=features_input, outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_df[['zero_cross_sum', 'zero_cross_rate_mean','mfcc_mean_globalmfcc1_mean']], train_df['target'], epochs=10, batch_size=32)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler

# Load the song data into a DataFrame
songs_df = pd.read_csv("/content/gdrive/MyDrive/songinfos (2).csv")

# Extract the features of interest (zcr, chroma, mfcc)
features = ['zero_cross_sum','zero_cross_rate_mean','chromagram_mean']
song_features = songs_df[features]

# Scale the features between 0 and 1
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(song_features)

# Calculate the pairwise cosine similarity between the songs
similarity_matrix = cosine_similarity(scaled_features)

# Define a function that takes in a song title and returns the top 2 recommended songs
def recommend_songs(name, similarity_matrix, song_df, n_recommendations=2):
    # Get the index of the input song
    name_index =song_df.loc[song_df['name'] == name].index[0]
    # Get the similarity scores for the input song
    similarities = similarity_matrix[name_index]
    # Sort the songs by similarity score
    sorted_songs = song_df.iloc[np.argsort(similarities)[::-1]]
    # Get the top n recommended songs
    top_n = sorted_songs.iloc[1:n_recommendations+1]
    # Return the title of the recommended songs
    return top_n['name']

# Test the function by recommending songs for a sample song
print(recommend_songs("Sample Song Title", similarity_matrix, songs_df))
```
